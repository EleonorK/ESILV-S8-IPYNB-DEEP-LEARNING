Creation d'une def pour afficher plusieur images (petit nombre)

```python
def displayImages(nbImages,imageDataset, labels):
   i= 0
   for image in imageDataset: 
          plt.figure()
          plt.imshow (imageDataset[i])
          plt.colorbar()
          plt.grid ( False )
          plt.xticks([])
          plt.yticks([])
          plt.xlabel(labels[i])
          plt.show()
          i = i + 1
          if i == nbImages:
             break
displayImages(2,train_images, train_labels)
```

Creation d'une def pour afficher plusieur images dans un grid (grand nombre)

```python
def displayGridImages(nbImages,imageDataset, labels, colorMap):

  plt.figure(figsize=(28,28))
  for i in range(nbImages):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(imageDataset[i], cmap = colorMap)
    plt.colorbar() 
    plt.xlabel(labels[i])
  plt.show()
displayGridImages(15,train_images,train_labels, plt.cm.bone)

#colormpas : 'binary', 'gist_yarg', 'gist_gray', 'gray', 'bone',
# 'pink', 'spring', 'summer', 'autumn', 'winter', 'cool','Wistia',
#'hot', 'afmhot', 'gist_heat', 'copper'
```

```python
#number of labels=number of classes
num_classes = 10
#define the shape of input data
input_shape = (28, 28, 1)
# Normalization:  Scale images to the [0, 1] range 
x_train = train_images.astype("float32") / 255
x_test = test_images.astype("float32") / 255
# Make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
#### ou ###
#x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))  # adapt this if using `channels_first` image data format
#x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))  # adapt this if using `channels_first` image data format
###########
print("x_train shape:", x_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")
# convert class arrays to binary class matrices
y_train = keras.utils.to_categorical(train_labels, num_classes)
y_test = keras.utils.to_categorical(test_labels, num_classes)
```

#Exercise 3: A simple Neural network model to image classification

Prepare the data
  1. normalization
  2. check if dataset is of the shape of 28x28x1

```python
from keras import layers
model1= keras.Sequential([
  layers.Flatten(input_shape=input_shape), #prendre la mat et la mettre en 1 ligne
    layers.Dense(128, activation='relu'), #pwer of 2 so that it easy to don't remember
    layers.Dense(10, activation='softmax')
])
model1.summary()
```

```python
from keras import layers
model2 = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ]
)
model2.summary()
```

model parameters optimization

```python
model1.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model2.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

train model => model1

```python
batch_size = 128
epochs = 15
model1.fit(x_train,y_train,batch_size=batch_size,epochs=epochs, validation_split = 0.2)
```

train model => model2

```python
batch_size = 128
epochs = 15
model2.fit(x_train,y_train,batch_size=batch_size,epochs=epochs, validation_split = 0.2)
```

#Exercise 5: Model Evaluation

```python
score2 = model2.evaluate(x_test, y_test, verbose=2)
print("Test loss:", score2[0])
print("Test accuracy:", score2[1])
```

```python
score1 = model1.evaluate(x_test, y_test, verbose=2)
print("Test loss:", score1[0])
print("Test accuracy:", score1[1])
```
